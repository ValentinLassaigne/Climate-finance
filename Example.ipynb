```python
def portfolio_return(weights,stocks):
    return(pd.DataFrame(((weights*stocks).transpose()).sum(),index = index.index))
```

```python
from scipy.optimize import minimize
    
def mTE(stocks,indx):
    """
    Returns the weights of the portfolio that gives you the maximum sharpe ratio
    given the riskfree rate and expected returns and a covariance matrix
    """
    n = stocks.shape[1]
    T = stocks.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),) * n # an N-tuple of 2-tuples!
    # construct the constraints
    weights_sum_to_1 = {'type': 'eq',
                        'fun': lambda weights: np.sum(weights) - 1
    }
    def TE(weights, stocks, indx):
        """
        Returns the tracking error
        """
        r = portfolio_return(weights, stocks)
        diff = r.values - indx.values
        return ((diff)**2/T).sum()
    
    weights = minimize(TE, init_guess,
                       args=(stocks, indx), method='SLSQP',
                       options={'disp': False},
                       constraints=(weights_sum_to_1,),
                       bounds=bounds)
    return weights.x
```

# Strategy 1 : max (ENC)

```python
from scipy.optimize import minimize
    
def mENC(stocks,indx):
    """
    Returns the weights of the portfolio that gives you the maximum sharpe ratio
    given the riskfree rate and expected returns and a covariance matrix
    """
    n = stocks.shape[1]
    T = stocks.shape[0]
    #init_guess = np.repeat(1/n, n)
    init_guess = np.repeat(1/n,n)
    bounds = ((0.0, 1.0),) * n # an N-tuple of 2-tuples!
    # construct the constraints
    weights_sum_to_1 = {'type': 'eq',
                        'fun': lambda weights: np.sum(weights) - 1
    }
    def TE(weights, stocks, indx):
        """
        Returns ENC
        """
        #return -1/sum(weights[i]**2 for i in range(len(weights)))
        return -1/(((weights)**2).sum())
    weights = minimize(TE, init_guess,
                       args=(stocks, indx), method='SLSQP',
                       options={'disp': False},
                       constraints=(weights_sum_to_1,),
                       bounds=bounds)
    return weights.x


```

```python
from scipy.optimize import minimize
    

def portfolio_return(weights,stocks):
    return(pd.DataFrame(((weights*stocks).transpose()).sum(),index = index.index))


def mTE2(stocks,indx):
    """
    Returns the weights of the portfolio that gives you the maximum sharpe ratio
    given the riskfree rate and expected returns and a covariance matrix
    """
    n = stocks.shape[1]
    T = stocks.shape[0]
    #init_guess = np.repeat(1/n, n)
    init_guess = np.zeros(n)
    init_guess[0]=0.5
    init_guess[1]=0.5
    bounds = ((0.0, 1.0),) * n # an N-tuple of 2-tuples!
    # construct the constraints
    weights_sum_to_1 = {'type': 'eq',
                        'fun': lambda weights: np.sum(weights) - 1
    }
    def TE(weights, stocks, indx):
        """
        Returns the tracking error
        """
        r = portfolio_return(weights, stocks)
        diff = r.values - indx.values
        #print(((diff)**2/T).sum())
        return ((diff)**2/T).sum()
    
    weights = minimize(TE, init_guess,
                       args=(stocks, indx), method='SLSQP',
                       options={'disp': False},
                       constraints=(weights_sum_to_1,),
                       bounds=bounds)
    return weights.x


```

```python
from scipy.optimize import minimize
    

def portfolio_return(weights,stocks):
    return(pd.DataFrame(((weights*stocks).transpose()).sum(),index = index.index))


def mVAR(stocks,indx):
    """
    Returns the weights of the portfolio that gives you the maximum sharpe ratio
    given the riskfree rate and expected returns and a covariance matrix
    """
    n = stocks.shape[1]
    T = stocks.shape[0]
    #init_guess = np.repeat(1/n, n)
    init_guess = np.zeros(n)
    init_guess[0]=0.5
    init_guess[1]=0.5
    bounds = ((0.0, 1.0),) * n # an N-tuple of 2-tuples!
    # construct the constraints
    weights_sum_to_1 = {'type': 'eq',
                        'fun': lambda weights: np.sum(weights) - 1
    }
    def VAR(weights, stocks, indx):
        """
        Returns the variance
        """
        return ((weights*Var)**2).sum()
    
    weights = minimize(VAR, init_guess,
                       args=(stocks, indx), method='SLSQP',
                       options={'disp': False},
                       constraints=(weights_sum_to_1,),
                       bounds=bounds)
    return weights.x


```

```python
w=mVAR(stocks2,index)
plt.plot(w)
```
